the continuous images we are getting we are converting them to greyscake and flipping it+ black and white+ cropped.

```python
vid= cv2.VideoCapture(0)

while True:
    flag, img= vid.read()
    if flag:
        img_gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        th, img_binary= cv2.threshold(img_gray, 100, 255, cv2.THRESH_BINARY)

        x1, y1, w, h= (200, 400, 200, 200)
        img_cropped= img[y1: y1+h, x1: x1+w, :]  

        cv2.rectangle(
            img
            pt1= (x1+w, y1+h))      


        cv2.imshow('Prieview', img_cropped)
        Key= cv2.waitKey(1)
        if Key== ord('x'):
            break
    else:
        break
    sleep(0.1)
vid.release()        
cv2.destroyAllWindows()
```
