```python
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('desktop-wallpaper-beautiful-pastel-aesthetic-minimalist-macbook.jpg')
img2= cv2.imread('EEfry4XU4AAFxQt.jpg')
img2= img2[150:150+288, 150:150+288,:]
```

```python
img.shape, img2.shape
```

```python
def showing(im):
    plt.imshow(im[:,:,::-1])
    plt.show()

def imshow(im):
    cv2.imshow('preview', im)
    cv2.waitKey(0)
    cv2.dastroyAllWindows()
```

```python
imshow(img)
imshow(img2)
```

```python
showing(img)
```
